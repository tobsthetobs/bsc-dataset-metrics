```python
# Imports
import os
import numpy as np
import sys
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
dataset_folder = 'Datasets'

def count_kitti():
    # Setup directories using os 
    img_folder = 'KITTI/dataset/sequences'
    cur_dir = os.path.normpath(os.getcwd() + os.sep + os.pardir)
    dir = cur_dir + "/" + dataset_folder + "/" + img_folder
    data = os.listdir(dir)
    
    # Setup lists and variables
    counter = 0
    sum =  0
    index = -1
        
    # Iterate over folders
    for sequencefolder in data:
        for subfolder in os.listdir(os.path.join(dir, sequencefolder)): 
            index += 1
            for file in os.listdir(os.path.join(dir, sequencefolder, subfolder)):
                counter += 1
            print("KITTI sequence: ", index, " with: ",counter)
            sum += counter
            # print("Scanning next folder current total of images processed: ", sum)
            counter = 0
    return 
count_kitti()
```

```python
import tkinter as tk
from tkinter import filedialog

def select_file():
    root = tk.Tk()
    root.withdraw()
    file_path = filedialog.askopenfilename()
    return file_path
def select_dir():
    root = tk.Tk()
    root.withdraw()
    dir_path = filedialog.askdirectory()
    return dir_path
```

```python
curdir = os.path.normpath(os.getcwd() + os.sep + os.pardir)
high_entropy = (pd.read_csv(curdir + '/Tensorboard_files/saved/train.csv'),pd.read_csv(curdir + '/Tensorboard_files/saved/validation.csv'))
```

```python
low_entropy = (pd.read_csv(curdir + '/Tensorboard_files/saved/train_low.csv'),pd.read_csv(curdir + '/Tensorboard_files/saved/validation_low.csv'))
```

```python
step = 0
f, ax = plt.subplots()
for i in low_entropy:
    if step == 0:
        #i.plot(x='Step', y='Value', title="Training", legend=True, ylabel="Epoch Loss",grid=True)
        ax.plot(i["Step"], i["Value"], label='Training')
    else: 
        #i.plot(x='Step', y='Value', title="Validation", legend=True, ylabel="Epoch Loss",grid=True)
        ax.plot(i["Step"], i["Value"],label='Validation')
    step += 1
plt.legend()
plt.style.use('fivethirtyeight')
plt.xlabel('Step')
plt.ylabel('Value')
plt.show()
```
